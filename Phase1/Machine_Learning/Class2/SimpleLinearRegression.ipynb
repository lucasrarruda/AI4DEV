```python
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

data_frame = pd.read_csv('database.csv')
data_frame.head()
```

```python
plt.scatter(data_frame['age'], data_frame['salary'])
plt.xlabel('Age')
plt.ylabel('Salary (USD)')
plt.title('Age vs Salary')
plt.show()
```

```python
data_frame.corr()
```

```python
data_x = data_frame['age']
data_y = data_frame['salary']

x_train, x_test, y_train, y_test = train_test_split(data_x, data_y, test_size=0.2, random_state=42)

x_train = x_train.values.reshape(-1, 1)
x_test = x_test.values.reshape(-1, 1)
```

```python
model = LinearRegression()
model.fit(x_train, y_train)

prediction = model.predict(x_test)
```

```python
mse = mean_squared_error(y_test, prediction)
mae = mean_absolute_error(y_test, prediction)
r2 = r2_score(y_test, prediction)

print(f'Mean Squared Error: {mse}')
print(f'Mean Absolute Error: {mae}')
print(f'R2 Score: {r2}')
```

```python
plt.scatter(x_test, y_test, color='blue', label='Real')
plt.scatter(x_test, prediction, color='red', label='Prediction')
plt.xlabel('Age')
plt.ylabel('Salary (USD)')
plt.title('Estimation: Age vs Salary')
plt.legend()
plt.show()
```
